### 1. Plot the GDP as a function of the year in Jupyter as shown.

```python
%matplotlib inline
import matplotlib.pyplot as graph

x = [1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
y = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599]
graph.plot(x, y, 'or--')
graph.title('Real US GDP (in trillions)')
graph.xlabel('Year')
graph.ylabel('Trillions of $')
graph.show()
```

### 2. Based on the obtained plot, what is a possible mathematical relationship between GDP and Year?

The relationship that stems from this data is that every 10 years, the GDP exponentially increases. Therefore, a polynomial equation exists for this relationship. The equation can be described by taking the average change over each decade and dividing by the total number of decades minus one (minus one because there are 8 differences).

$$\frac{Average\;change\;of\;GDP\ during\;each\; decade}{Number\;of\;Decades - 1}$$

So:
$$\frac{(1.33/1.015)+(2.29/1.33)+(3.26/2.29)+(4.951/3.26)+(6.759/4.951)+(9.366/6.759)+(13.131/9.366)+(15.599/13.131)}{(9-1)}$$

From this equation, we see that there as an average of an 1.4144 increase each year.

Ultimately, the  equation will be in some form of:
$$y = \;ax^2+bx+c\;$$




### 3. Plot another curve that follows the mathematical relationship you have derived/argued between GDP and Year. This plot should be overlaid on top of the previous plot obtained using the actual GDP data.

```python
%matplotlib inline
import matplotlib.pyplot as graph

x = [1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010]
y = [1.015, 1.33, 2.29, 3.26, 4.951, 6.759, 9.366, 13.131, 15.599]
arr = []
for i in range(len(y)):
    deriv = (1.4144*y[i])
    arr.append(deriv)


graph.plot(x, y,'or--')
graph.plot(x,arr,'ob--')
graph.title('Real US GDP (in trillions)')
graph.xlabel('Year')
graph.ylabel('Trillions of $')
graph.show()
```
